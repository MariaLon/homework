Задача 1.
Напишите программу, которая будет проверять тексты на попытки обмануть антиплагиат с помощью подмены некоторых символов на такие же, но латинские. Программа должна открывать текст в формате .txt, разделять его на токены любым способом и проверять каждый токен. Если токен содержит только латинские или кириллические символы, то с ним все хорошо, а если в нем есть 1-2 латинских символа и все остальные - кириллические, то программа должна печатать этот токен и номер его строки с указанием латинских букв. Строки можно нумеровать с помощью счетчика, когда читаете файл. Лучше всего считывать файл построчно, очевидно.

```python
chLat = ''.join([chr(x) for x in range(ord('a'), ord('z'))]) 
chLat += chLat.upper()
chCyr = ''.join([chr(x) for x in range(ord('а'), ord('я'))]) 
chCyr += chCyr.upper() +'ёЁ'
punctuations = '.,~+-_/\|*^=@#$%&`'+"'"+'"!?:;<>{}[]()'

def TestWord(Word):
    Lat = False
    Cyr = False
    for ch in Word:
        if ch in chLat:
            Lat = True
        if ch in chCyr:
            Cyr = True
        if Lat & Cyr:
            break
    return Lat & Cyr

fname = input("File Path:")       
Line = 0
for fl in open(fname, 'r', encoding='utf8'):
    Line += 1
    Word = 0
    for W in (''.join([' ' if x in punctuations else x for x in fl])).split():
        Word += 1
        if TestWord(W):
            print(f'Line:{Line} Word:{Word} [{W}] Lat.chars:{",".join([x for x in W if x in chLat])}')
```

Напишите программу для игры в города. При загрузке программа должна подгружать файл с названиями городов (вам придется его составить, можно поискать готовые списки). Сама игра происходит следующим образом: начинает игрок-человек и вводит любое название города. Программа проверяет, на какую букву заканчивается это название, и случайным образом выбирает подходящий город на эту букву в своем списке (для случайного выбора можно использовать библиотеку random и функцию choice). Если город заканчивается на 'ь', берем предпоследнюю букву (или можно играть по-английски). Когда человек вводит следующий ответ, программа должна проверить, начинается ли введенный город на правильную букву, и если нет, то потребовать ввести заново. Чтобы остановить игру, человек должен просто нажать Enter. Не забудьте создать множество для уже введенных городов, ведь повторяться нельзя! (Список городов у программы тоже на самом деле может быть множеством или даже словарем, чтобы удобнее было по буквам отсеивать).

+1 дополнительный балл, если напишете в классах.

```python
import random
Alphabet = 'абвгдежзийклмнопрстуфхцчшщъыьэюяАБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯёЁ'

class CitiesGame:
  def __init__(self):
    self._spent = []
    self._C_turn = False
    with open('cities.txt') as f:
      self._cities = [c[:-1] for c in f.readlines()]
  
  def _LastCh(self):
    out = ''
    for c in self._spent[-1][::-1]:
      if not c in 'ьЬ':
        out = c
        break
    return out

  def _Accordance(self, city):
    if len(self._spent) > 0:
      return self._LastCh() == city[0].lower()
    else:
      return True

  def _Check(self, city):
    return len(city) > 0 and city[0] in Alphabet and city[-1] in Alphabet

  def _turn(self):
    out = False
    if self._C_turn:
      tmp = list(filter(lambda c: c[0] == self._LastCh().upper() and not c in self._spent, self._cities))
      if len(tmp) > 0:
        self._spent.append(random.choice(tmp))
        print('- ' + self._spent[-1])
        out = True
      else:
        print('Вы победили!')
    else:
      while True:
        city = input().title()
        if city in self._spent:
          print('Этот город уже был. Попробуйте ещё...')
        elif self._Check(city):
          if self._Accordance(city):
            self._spent.append(city)
            out = True
            break
          else:
            print(f'Название города не на букву "{self._LastCh()}"!')
            print('Попробуйте ещё...')
        else:
          if city == '':
            print('Игра остановлена.')
            break
          else:
            print('Неправильный ввод. Поробуйте ещё...')
    self._C_turn = not self._C_turn
    return out

  def Play(self):
    print('Введите название города:')
    while True:
      if not self._turn():
        self._C_turn = False
        break;
    self._spent = []

game = CitiesGame()
game.Play()

```
