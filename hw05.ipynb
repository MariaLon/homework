Задача 1.
Просмотрите оба выбранных текста. Удостоверьтесь, что тексты чистые, если же в них есть какой-то мусор: хештеги, затесавшиеся при OCR символы и подобное, почистите с помощью регулярных выражений.

Проведите первичный статистический анализ: разбейте тексты на предложения и на токены, посчитайте относительное количество того и другого, сопоставьте. Если ваши тексты параллельные, какой длиннее? В каком тексте средняя длина предложения больше? Почему? В каком тексте выше лексическое разнообразие?

Таким образом, вам необходимо узнать следующие вещи:

количество предложений (относительное и абсолютное)
количество токенов (относительное и абсолютное)
средняя длина предложения (среднее количество слов в предложении)
соотношение "уникальные токены / все токены"
(опционально) соотношение знаков пунктуации и слов



```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
#!pip install pymorphy2
import pymorphy2


class TAnalizer:
  def __init__(self, Text, language="english"):
    morph = pymorphy2.MorphAnalyzer()
    Txt = [[[token, not 'PNCT' in morph.parse(token)[0].tag] for token in word_tokenize(sent, language)] for sent in sent_tokenize(Text, language)]
    self.SentCount = len(Txt)
    self.TokenCount = sum([len(sent) for sent in Txt])
    WordsCount = sum([len([token for token in sent if token[1]]) for sent in Txt]) 
    self.AverageWordsInSent = WordsCount / self.SentCount
    UniqueWordCount = len(set([token[0].title() for sent in Txt for token in sent if token[1]]))
    self.LexicalVariety = UniqueWordCount / WordsCount
    self.PnctWordsCorrelation = (self.TokenCount - WordsCount) / WordsCount

  @classmethod
  def fromFile(cls, FileName, language="english"):
    file = open(FileName)
    return cls(file.read(), language)

  def __repr__(self):
    return "TAnalizer()"
  def __str__(self):
    return 'Sent Count:' + str(self.SentCount) + \
    '\nTokens Count:' + str(self.TokenCount) + \
    '\nAverage Words In Sent:' + str(self.AverageWordsInSent) + \
    '\nLexical Variety:' + str(self.LexicalVariety) + \
    '\nPnctuation and Words Correlation:' + str(self.PnctWordsCorrelation)
    

RusT = TAnalizer.fromFile('text-1.txt', 'russian')
SpnT = TAnalizer.fromFile('text-2.txt', 'spanish')

print(f'{"Rus":20}|{"Spn":20}')
print('Абсолютное количество предложений')
print(f'{RusT.SentCount:20}|{SpnT.SentCount:20}')
print('Относительное количество предложений')
sm = RusT.SentCount + SpnT.SentCount
print(f'{RusT.SentCount/sm:20}|{SpnT.SentCount/sm:20}')
print('Абсолютное количество токенов')
print(f'{RusT.TokenCount:20}|{SpnT.TokenCount:20}')
print('Относительное количество токенов')
sm = RusT.TokenCount + SpnT.TokenCount
print(f'{RusT.TokenCount/sm:20}|{SpnT.TokenCount/sm:20}')
print('средняя длина предложения')
print(f'{RusT.AverageWordsInSent:20}|{SpnT.AverageWordsInSent:20}')
print('Лексическое разнообразие')
print(f'{RusT.LexicalVariety:20}|{SpnT.LexicalVariety:20}')
print('соотношение знаков пунктуации и слов')
print(f'{RusT.PnctWordsCorrelation:20}|{SpnT.PnctWordsCorrelation:20}')
```

Сделайте морфосинтаксические разборы ваших текстов в формате UD, запишите .conllu-файлы.

```python
#!pip install spacy_udpipe
import spacy_udpipe
#spacy_udpipe.download('ru')
#spacy_udpipe.download('es')

class TAnalizer:
  def __init__(self, Text, language="en"):
    nlp= spacy_udpipe.load(language)
    doc = nlp(Text)
    self.Tokens = [[token.i, token.text, token.lemma_, \
     token.pos_, token.dep_, token.head.text] for token in doc]

  @classmethod
  def fromFile(cls, FileName, language="en"):
    file = open(FileName)
    return cls(file.read(), language)
  
  def print(self):
    for token in self.Tokens:
      print(f'{token[0]:2}. {token[1]:15} Lemma: {token[2]:15} \
POS: {token[3]:6} SyntR: {token[4]:9} Head: {token[5]}')

  def __repr__(self):
    return "TAnalizer()"
  def __str__(self):
    return self.Tokens
    

RusT = TAnalizer.fromFile('text-1.txt', 'ru')
SpnT = TAnalizer.fromFile('text-2.txt', 'es')

print(f'Russian Text\n')
RusT.print()
print(f'\nSpanish Text\n')
SpnT.print()
```

Задача 3.
Посчитайте статистику по частям речи, сопоставьте: можно напечатать две таблички с процентами по частям речи.

Count POS statistics for both texts: there should be two tables with POS-tags and their quantity in the text, e.g. NOUN 23%.

```python
from nltk.corpus.reader.wordnet import itemgetter
#!pip install spacy_udpipe
import spacy_udpipe
#spacy_udpipe.download('ru')
#spacy_udpipe.download('es')

class TAnalizer:
  def __init__(self, Text, language="en"):
    nlp= spacy_udpipe.load(language)
    doc = nlp(Text)
    POS = [token.pos_ for token in doc]
    self.POS = sorted([[tag, \
     len(list(filter(lambda x: x == tag, POS))) / len(POS)] \
     for tag in set(POS)], key=itemgetter(0))

  @classmethod
  def fromFile(cls, FileName, language="en"):
    file = open(FileName)
    return cls(file.read(), language)

  def __repr__(self):
    return "TAnalizer()"
  def __str__(self):
    return "\n".join([f'{l[0]:6}: {l[1]*100:20}%' for l in self.POS])
    

RusT = TAnalizer.fromFile('text-1.txt', 'ru')
SpnT = TAnalizer.fromFile('text-2.txt', 'es')
print(f'Russian Text\n{RusT}\n')
print(f'Spanish Text\n{SpnT}')
```

```python
from nltk.corpus.reader.wordnet import itemgetter
#!pip install spacy_udpipe
import spacy_udpipe
#spacy_udpipe.download('ru')
#spacy_udpipe.download('es')

class TAnalizer:
  def __init__(self, Text, language="en"):
    nlp= spacy_udpipe.load(language)
    doc = nlp(Text)
    Tokens = [[token.text, token.lemma_, token.pos_] for token in doc]
    self.POS = sorted([[tag, \
     len(list(filter(lambda x: x[2] == tag and x[0] == x[1], Tokens))) / \
     len(list(filter(lambda x: x[2] == tag, Tokens)))] 
     for tag in set([i[2] for i in Tokens])], key=itemgetter(0))

  @classmethod
  def fromFile(cls, FileName, language="en"):
    file = open(FileName)
    return cls(file.read(), language)

  def __repr__(self):
    return "TAnalizer()"
  def __str__(self):
    return "\n".join([f'{l[0]:6}: {(1 - l[1]) * 100:20}%' for l in self.POS])
    

RusT = TAnalizer.fromFile('text-1.txt', 'ru')
SpnT = TAnalizer.fromFile('text-2.txt', 'es')
print(f'Russian Text\n{RusT}\n')
print(f'Spanish Text\n{SpnT}')
```
